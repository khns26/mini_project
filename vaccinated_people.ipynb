```
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
```

```
my_filepath = "country_vaccinations.csv"
my_data = pd.read_csv(my_filepath)
my_data
```

```
data = [my_data["country"], my_data["date"], my_data["people_vaccinated"]]
df = pd.DataFrame(data)
df = df.transpose()
df.fillna(0, inplace=True)
df = df.groupby(['country']).sum()
df
```

```
df = df.groupby(['country']).sum()# Your code here
df
df.plot.bar(figsize=(20,10))
plt.title("Vaccinated People in Each Country")
plt.xlabel("Country")
plt.xticks(rotation=90)
plt.ylabel("Number of Vaccinated People")
plt.show()
```
